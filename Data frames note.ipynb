## Visualizaci√≥n de DataFrames en Pandas

- **`head()`**

### Filtrar Data
El m√©todo **`loc`** nos permite filtrar filas de un DataFrame usando condiciones l√≥gicas y seleccionar columnas espec√≠ficas.

---

### üìè Filtrar por una condici√≥n simple

```python
bios.loc[bios['height_cm'] > 215, ['name', 'height_cm']]



````python
bios.loc[bios['height_cm'] > 215, ['name', 'height_cm']]
#multiples condiciones
bios.loc[(bios['height_cm'] > 215) & (bios['born_country'] == 'USA'), ['name', 'height_cm', 'born_country']]
#Si queremos ver solo los nombres que empezan con la letra C
bios.loc[bios['name'].str.contains('Carlo'), ['name']]
````
